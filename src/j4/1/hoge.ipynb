```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# CSV
df = pd.read_csv("execution_times_mycp.csv")

# 
plt.figure(figsize=(10, 6))

# 
plt.plot(df['size'], df['real_time'], label='Real Time', marker='o')
plt.plot(df['size'], df['user_time'], label='User Time', marker='s')
plt.plot(df['size'], df['sys_time'], label='System Time', marker='^')

# 
plt.xlabel('Size')
plt.ylabel('Time (s)')
plt.title('Execution Times by Size')

# 
plt.legend()

# 
plt.xscale('log',base=16)
plt.ylim(0, 40)

# 
plt.grid(True, which="major", ls="--")

# 
plt.show()
```
